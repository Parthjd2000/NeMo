```python
import re
from datetime import datetime

# ---- Dialogue state you can persist across turns ----
STATE = {
    "name": None,
    "school": None,
    "is_graduate_student": False,
    "last_intent": None,
}

# ---- Simple text cleanup ----
def normalize_text(t: str) -> str:
    t = t.strip()
    # keep original for UI; use lower for matching
    return t, t.lower()

# ---- Intent rules ----
INTENTS = {
    "greet": [r"\b(hi|hello|hey)\b"],
    "self_intro": [r"\bmy name is\b", r"\bi am\b", r"\bi'm\b"],
    "help_request": [r"\b(can you|could you|help|how do i|i need)\b"],
    "thanks": [r"\b(thanks|thank you|appreciate it)\b"],
    "goodbye": [r"\b(bye|goodbye|see you)\b"],
}

def detect_intent(text_lc: str) -> str:
    for intent, patterns in INTENTS.items():
        for p in patterns:
            if re.search(p, text_lc):
                return intent
    return "unknown"

# ---- Slot/Entity extraction ----
def extract_entities(text: str, text_lc: str):
    # name
    name = None
    m = re.search(r"\bmy name is\s+([A-Za-z][A-Za-z\s\.'-]{1,60})", text, flags=re.I)
    if not m:
        # fallback: "I am <name>"
        m = re.search(r"\b(i am|i'm)\s+([A-Za-z][A-Za-z\s\.'-]{1,60})", text, flags=re.I)
        if m: name = m.group(2)
    else:
        name = m.group(1)

    if name:
        name = " ".join(w.capitalize() for w in name.strip().split())

    # school / org after "at ..."
    school = None
    m = re.search(r"\bat\s+([A-Za-z0-9&\.\-'\s]{2,80})", text, flags=re.I)
    if m:
        school = m.group(1).strip(" .,")

    # grad student
    is_grad = bool(re.search(r"\bgraduate student\b", text_lc))

    return {"name": name, "school": school, "is_grad": is_grad}

def update_state(STATE, ents):
    if ents["name"]: STATE["name"] = ents["name"]
    if ents["school"]: STATE["school"] = ents["school"]
    if ents["is_grad"]: STATE["is_graduate_student"] = True

# ---- Response generation (templates) ----
def generate_reply(intent: str, user_text: str, STATE):
    name = STATE.get("name")
    school = STATE.get("school")
    is_grad = STATE.get("is_graduate_student")
    polite = "How can I help you today?"

    if intent == "greet":
        return f"Hi{f' {name}' if name else ''}! {polite}"

    if intent == "self_intro":
        bits = []
        if name: bits.append(f"Nice to meet you, {name}!")
        else: bits.append("Nice to meet you!")
        if is_grad and school: bits.append(f"That’s awesome you’re a graduate student at {school}.")
        elif is_grad:          bits.append("That’s awesome you’re a graduate student.")
        elif school:           bits.append(f"Hope things are going well at {school}.")
        bits.append(polite)
        return " ".join(bits)

    if intent == "help_request":
        return ("Sure—tell me a bit more about what you’d like to do. "
                "If it’s ASR, I can help you transcribe files, batch a folder, or export to CSV.")

    if intent == "thanks":
        return "You’re welcome! If you want, I can also optimize your pipeline or evaluate accuracy (WER)."

    if intent == "goodbye":
        return f"Bye{f', {name}' if name else ''}! Have a great day."

    # unknown → safe, helpful default
    return f"Got it. {polite}"

# ---- One function to go from ASR result → reply text ----
def reply_to_asr(nemo_result, state=STATE):
    # Handle both plain strings and NeMo Hypothesis objects
    user_text = getattr(nemo_result, "text", str(nemo_result))
    text, text_lc = normalize_text(user_text)

    intent = detect_intent(text_lc)
    ents = extract_entities(text, text_lc)
    update_state(state, ents)
    state["last_intent"] = intent

    reply = generate_reply(intent, text, state)
    return user_text, intent, ents, reply

# ---------- Demo with your sentence ----------
asr_text = "My name is Parth Desai, I'm a graduate student at SDSU"
user_text, intent, ents, bot_reply = reply_to_asr(asr_text, STATE)
print("ASR:", user_text)
print("Intent:", intent)
print("Entities:", ents)
print("Bot:", bot_reply)
```

```python
# ---------- 0) Setup: tiny LLM (only used on fallback) ----------
!pip install -q transformers accelerate sentencepiece

import re
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tok = AutoTokenizer.from_pretrained("google/flan-t5-small")
llm = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small", device_map="auto")

# ---------- 1) Minimal rule-based engine ----------
STATE = {"name": None, "school": None, "is_graduate_student": False, "last_intent": None}

def normalize_text(t: str):
    return t.strip(), t.strip().lower()

INTENTS = {
    "greet":       [r"\b(hi|hello|hey)\b"],
    "self_intro":  [r"\bmy name is\b", r"\bi am\b", r"\bi'm\b"],
    "help_request":[r"\b(can you|could you|help|how do i|i need)\b"],
    "thanks":      [r"\b(thanks|thank you|appreciate it)\b"],
    "goodbye":     [r"\b(bye|goodbye|see you)\b"],
}

def detect_intent(text_lc: str) -> str:
    for intent, patterns in INTENTS.items():
        if any(re.search(p, text_lc) for p in patterns):
            return intent
    return "unknown"

def extract_entities(text: str, text_lc: str):
    # name
    name = None
    m = re.search(r"\bmy name is\s+([A-Za-z][A-Za-z\s\.'-]{1,60})", text, flags=re.I)
    if not m:
        m = re.search(r"\b(i am|i'm)\s+([A-Za-z][A-Za-z\s\.'-]{1,60})", text, flags=re.I)
        if m: name = m.group(2)
    else:
        name = m.group(1)
    if name:
        name = " ".join(w.capitalize() for w in name.strip().split())

    # school (optional; not used in your two tests but kept here)
    school = None
    m = re.search(r"\bat\s+([A-Za-z0-9&\.\-'\s]{2,80})", text, flags=re.I)
    if m:
        school = m.group(1).strip(" .,")

    is_grad = bool(re.search(r"\bgraduate student\b", text_lc))
    return {"name": name, "school": school, "is_grad": is_grad}

def update_state(state, ents):
    if ents["name"]: state["name"] = ents["name"]
    if ents["school"]: state["school"] = ents["school"]
    if ents["is_grad"]: state["is_graduate_student"] = True

def generate_reply(intent: str, state):
    name = state.get("name")
    polite = "How can I help you today?"
    if intent == "greet":
        return f"Hi{f' {name}' if name else ''}! {polite}"
    if intent == "self_intro":
        return f"Nice to meet you{f', {name}' if name else ''}! {polite}"
    if intent == "help_request":
        return "Sure—tell me what you need help with."
    if intent == "thanks":
        return "You’re welcome!"
    if intent == "goodbye":
        return f"Bye{f', {name}' if name else ''}!"
    return None  # unknown → trigger LLM fallback

# ---------- 2) Tiny LLM fallback ----------
SYSTEM = (
    "You are a concise, friendly assistant. "
    "Reply in 1–2 sentences. Use the user's name if provided."
)

def llm_reply(user_text, state, max_new_tokens=64):
    ctx = []
    if state.get("name"): ctx.append(f"UserName: {state['name']}")
    if state.get("school"): ctx.append(f"UserSchool: {state['school']}")
    if state.get("is_graduate_student"): ctx.append("UserIsGrad: true")
    ctx = "\n".join(ctx)

    prompt = f"{SYSTEM}\n{ctx}\nUser: {user_text}\nAssistant:"
    inp = tok(prompt, return_tensors="pt").to(llm.device)
    out = llm.generate(**inp, max_new_tokens=max_new_tokens, do_sample=True, top_p=0.9, temperature=0.6)
    return tok.decode(out[0], skip_special_tokens=True).strip()

# ---------- 3) Hybrid policy function ----------
def smart_reply(user_text: str):
    text, text_lc = normalize_text(user_text)
    intent = detect_intent(text_lc)
    ents = extract_entities(text, text_lc)
    update_state(STATE, ents)

    # Try rules first
    rule_resp = generate_reply(intent, STATE)
    if rule_resp:
        return {"input": user_text, "path": "rule-based", "intent": intent, "entities": ents, "reply": rule_resp}

    # Fallback to tiny LLM
    llm_resp = llm_reply(user_text, STATE)
    return {"input": user_text, "path": "tiny-llm (FLAN-T5-small)", "intent": intent, "entities": ents, "reply": llm_resp}

# ---------- 4) Test your two sentences ----------
tests = [
    "Hello, my name is Parth Desai",
    "What is the weather today?"
]

results = [smart_reply(t) for t in tests]
for r in results:
    print("──────────")
    print("Input: ", r["input"])
    print("Path:  ", r["path"])
    print("Intent:", r["intent"])
    print("Entities:", r["entities"])
    print("Reply:", r["reply"])
```

```python
# =========================
# 0) Imports & tiny helpers
# =========================
!apt-get -y update >/dev/null && apt-get -y install ffmpeg libsndfile1 >/dev/null

import os, re, subprocess, uuid, torch, soundfile as sf
from IPython.display import Audio, display

# NeMo ASR/TTS
import nemo.collections.asr as nemo_asr
from nemo.collections.tts.models import FastPitchModel, HifiGanModel

# Tiny LLM (fallback)
!pip -q install transformers accelerate sentencepiece >/dev/null
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# =========================
# 1) Load / prepare models
# =========================

# ---- ASR model (use your existing one if already loaded) ----
try:
    asr_model
except NameError:
    asr_model = nemo_asr.models.EncDecCTCModel.from_pretrained("QuartzNet15x5Base-En")

# ---- TTS models ----
fastpitch = FastPitchModel.from_pretrained("tts_en_fastpitch")

# pick a valid vocoder name on your NeMo version
infos = HifiGanModel.list_available_models()
names = [m.pretrained_model_name for m in infos]
preferred = [
    "tts_en_hifigan",
    "tts_en_hifitts_hifigan_ft_fastpitch",
    "tts_en_lj_hifigan_ft_mixerttsx",
    "tts_en_lj_hifigan_ft_mixertts",
]
chosen_vocoder = next((n for n in preferred if n in names), names[0])
vocoder = HifiGanModel.from_pretrained(chosen_vocoder)

# ---- Tiny LLM (fallback) ----
tok = AutoTokenizer.from_pretrained("google/flan-t5-small")
llm = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small", device_map="auto")

# =========================
# 2) Rule-based reply bits
# =========================
STATE = {"name": None, "school": None, "is_graduate_student": False, "last_intent": None}

def _normalize_text(t): return t.strip(), t.strip().lower()

INTENTS = {
    "greet":       [r"\b(hi|hello|hey)\b"],
    "self_intro":  [r"\bmy name is\b", r"\bi am\b", r"\bi'm\b"],
    "help_request":[r"\b(can you|could you|help|how do i|i need)\b"],
    "thanks":      [r"\b(thanks|thank you|appreciate it)\b"],
    "goodbye":     [r"\b(bye|goodbye|see you)\b"],
}

def _detect_intent(text_lc):
    for intent, pats in INTENTS.items():
        if any(re.search(p, text_lc) for p in pats):
            return intent
    return "unknown"

def _extract_entities(text, text_lc):
    name = None
    m = re.search(r"\bmy name is\s+([A-Za-z][A-Za-z\s\.'-]{1,60})", text, flags=re.I)
    if not m:
        m = re.search(r"\b(i am|i'm)\s+([A-Za-z][A-Za-z\s\.'-]{1,60})", text, flags=re.I)
        if m: name = m.group(2)
    else:
        name = m.group(1)
    if name: name = " ".join(w.capitalize() for w in name.strip().split())

    school = None
    m = re.search(r"\bat\s+([A-Za-z0-9&\.\-'\s]{2,80})", text, flags=re.I)
    if m: school = m.group(1).strip(" .,")

    is_grad = bool(re.search(r"\bgraduate student\b", text_lc))
    return {"name": name, "school": school, "is_grad": is_grad}

def _update_state(state, ents):
    if ents["name"]: state["name"] = ents["name"]
    if ents["school"]: state["school"] = ents["school"]
    if ents["is_grad"]: state["is_graduate_student"] = True

def _rule_reply(intent, state):
    name = state.get("name")
    polite = "How can I help you today?"
    if intent == "greet":      return f"Hi{f' {name}' if name else ''}! {polite}"
    if intent == "self_intro": return f"Nice to meet you{f', {name}' if name else ''}! {polite}"
    if intent == "help_request": return "Sure—tell me what you need help with."
    if intent == "thanks":     return "You’re welcome!"
    if intent == "goodbye":    return f"Bye{f', {name}' if name else ''}!"
    return None  # unknown → LLM

SYSTEM = ("You are a concise, friendly assistant. Reply in 1–2 sentences. "
          "Use the user's name if provided. Ask for missing info if needed.")

def _llm_reply(user_text, state, max_new_tokens=64):
    ctx = []
    if state.get("name"): ctx.append(f"UserName: {state['name']}")
    if state.get("school"): ctx.append(f"UserSchool: {state['school']}")
    if state.get("is_graduate_student"): ctx.append("UserIsGrad: true")
    prompt = f"{SYSTEM}\n" + ("\n".join(ctx) + "\n" if ctx else "") + f"User: {user_text}\nAssistant:"
    inp = tok(prompt, return_tensors="pt").to(llm.device)
    out = llm.generate(**inp, max_new_tokens=max_new_tokens, do_sample=True, top_p=0.9, temperature=0.6)
    return tok.decode(out[0], skip_special_tokens=True).strip()

# =========================
# 3) Utilities
# =========================
def ensure_wav_16k_mono(in_path):
    """
    Returns a path to a guaranteed mono/16k WAV (re-encodes if needed).
    """
    out_path = f"/content/{uuid.uuid4().hex}.wav"
    cmd = ["ffmpeg", "-y", "-i", in_path, "-ac", "1", "-ar", "16000", "-sample_fmt", "s16", out_path]
    subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, check=True)
    return out_path

def extract_text_from_nemo(result_item):
    # NeMo may return Hypothesis or str; try .text, else cast to str
    return getattr(result_item, "text", str(result_item))

def nemo_tts_speak(text, out_path="nemo_reply.wav", sr=22050):
    with torch.no_grad():
        tokens = fastpitch.parse(text)
        mel = fastpitch.generate_spectrogram(tokens=tokens)
        audio = vocoder.convert_spectrogram_to_audio(spec=mel).cpu().numpy()[0]
        sf.write(out_path, audio, sr)
    return out_path

# =========================
# 4) ONE-CALL PIPELINE
# =========================
def run_voice_chat(audio_path, use_llm_fallback=True, speak=True):
    """
    End-to-end: audio -> ASR -> reply -> (TTS)
    Returns a dict with transcript, reply, and audio path (if speak=True).
    """
    if not os.path.exists(audio_path):
        raise FileNotFoundError(f"Audio file not found: {audio_path}")

    safe_wav = ensure_wav_16k_mono(audio_path)

    # ASR
    asr_out = asr_model.transcribe([safe_wav])
    transcript = extract_text_from_nemo(asr_out[0])

    # Reply (rules -> LLM fallback)
    text, text_lc = _normalize_text(transcript)
    intent = _detect_intent(text_lc)
    ents = _extract_entities(text, text_lc)
    _update_state(STATE, ents)
    reply = _rule_reply(intent, STATE)
    if reply is None and use_llm_fallback:
        reply = _llm_reply(text, STATE)
    if reply is None:
        reply = "I’m here to help. Could you tell me more?"

    # TTS
    audio_reply_path = None
    if speak:
        audio_reply_path = nemo_tts_speak(reply, out_path="/content/nemo_reply.wav")

    # Pretty print + inline audio
    print("🎧 ASR transcript:", transcript)
    print("🤖 Reply path:", "rule-based" if reply != None and intent != "unknown" else ("tiny LLM" if use_llm_fallback else "rule-based (default)"))
    print("💬 Bot reply:", reply)
    if speak and audio_reply_path:
        display(Audio(audio_reply_path))

    return {"transcript": transcript, "reply": reply, "reply_wav": audio_reply_path, "intent": intent, "entities": ents}
```
